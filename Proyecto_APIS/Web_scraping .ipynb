# Extrayendo una tabla de internet territorios dependientes 

```python
# Ahora para tarer una imagen 
url = 'https://es.wikipedia.org/wiki/Territorio_dependiente'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
soup
```

```python
#Como la lista es plegable tomamos el elemento 0 para hacer el dataframe 
table = soup.find_all('table',{'class':'wikitable sortable'})[0]
table 
```

```python

cols=table.find_all('th')
cols = [col.text.strip() for col in cols]
cols
```

```python

rows =table.find_all('tr')
rows = [row.text.strip().split('\n') for row in rows][1:]
rows
#rows = re.sub(r'\xa0','',rows)
```

```python
colnames = cols
data = rows

df = pd.DataFrame(data, columns=colnames)
df.head(10)
```

# Seleccionar solo las columnas que nos interesan

```python

df_t= df['Territorio']
df_p = df['Potencia administradora']
```

```python
dftp = pd.concat([df_t, df_p], axis=1)
```

```python
# en esta tabla podemos ver a que potencias pertenecen las colonias
# En el caso de Sàhara Occidental , la potencia es España 
dftp
```

```python
#guardando la tabla en csv tabular 
```

```python
dftp = dftp.to_csv('dftp.csv',sep= '\t', index = False )
```
