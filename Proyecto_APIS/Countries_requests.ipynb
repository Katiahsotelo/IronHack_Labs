# En esta sección en donde descargamos todas las librería. Obtenemos el contenido html en formato json y lo pasamos a un DataFrame para visualizarlo

```python

import requests 
import json
import os 
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
#Hacer el requests para visualizar los datos 
def beggining(url):
    page=requests.get(url)
    return page.json()
```

```python
df = beggining('https://restcountries.com/v2/all')
```

```python
# Creando el DataFrame
df =pd.DataFrame(df)
df
```

# Ahora que vemos que los datos nos son úitiles, hacemos un output al folder donde vamos a empezar a trabajar

```python

#!curl https://restcountries.com/v2/all> output.json
        
```

```python
doc = open('output.json')
data = json.load(doc)
df=pd.DataFrame(data) 
df
```

# Empezamos a seleccionar, renombrar y reorganizar las columnas que nos interesan 

```python
df.rename(columns={'name':'Países/Colonias',
                        'capital':'Capital','subregion':'Subregión','region':'Región','population':'Población','latlng':'Latitud/longitud','area':'KM^2','languages':'Idiomas','independent':'Independiente','gini':'Gini'},
               inplace=True)
```

```python
dff =df.drop(['alpha2Code','alpha3Code','callingCodes','altSpellings','timezones','borders','nativeName','numericCode', 'flags', 'currencies','translations',
       'flag', 'regionalBlocs', 'cioc','topLevelDomain','demonym','Idiomas'], axis=1)

    
```

```python
#Reorganizando un columna 'latln'
# Se cambio el nombre y se corrigio a mano el único error NAN

df_lat=pd.DataFrame(dff['Latitud/longitud'].to_list(),columns = ['Latitud', 'Longitud'])
dff1 = pd.concat([dff, df_lat], axis=1)
dff2=dff1.drop(['Latitud/longitud'],axis=1)
```

```python
#Renombrar las columnas 
dff2 = dff2.reindex(columns=['Países/Colonias','Capital','Región','Subregión','Latitud','Longitud','Población','KM^2','Independiente','Gini'])
dff2
```

# Con los datos organizados ahora podemos empezar el análisis. 
Lo que queremos es tener un contexto de la situación general de los territorios dependientes. 
Para esto creamos una tabla con las características de interés  

```python
#Tabla de Gini 
df_p = dff2['Países/Colonias']
df_s= dff2['Subregión']
df_pop=dff2['Población']
df_in= dff2['Independiente']
df_g = dff2['Gini']
df_r = dff2['Región']
df_l = dff2['Latitud']
```

```python
df_todos= pd.concat([df_p, df_s,df_r,df_pop,df_l,df_in,df_g], axis=1)
```

```python
comparativo = df_todos[df_todos['Independiente']== False]
comparativo 
```

### Pruebas ####
En esta parte nos interesa conocer acerca de los Territorios dependientes

```python
#Número de Territorios dependientes 
#son 48 territoriod dependientes en el mundo 
len(comparativo)
```

### Para la región de Europa 

```python
#Solo Gibraltar(territorio británico en ultramar) esta en el subregión sur 
df_europa = comparativo[comparativo['Región']== 'Europe']
df_europa
```

```python
#En dónde se encuentran estos territorios dependientes 
np.unique(comparativo['Región'].values)
```

```python
#Existen 7 territorios dependientes en Europa
len(df_europa)
```

```python
# Svalbard and Jan Mayen (territorio de Noruga) cuenta con la menor población y 
#se encuentra a una mayor latitud ¿coincidencia?

df_europa.sort_values(by=['Población'], ascending= True)
```

### Para visualizar el índice de desigualdad Gini

```python
# De los 250 Territorios, solamente se tiene el gini de 167
dff2['Gini'].count()
```
