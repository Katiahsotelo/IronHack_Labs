```python
customers=orders.groupby(orders['CustomerID']).sum()['amount_spent']
customers_df= pd.DataFrame(customers)
customers_df
```

```python
customers_df['Type_customer']=np.where((customers_df['amount_spent']>=customers.quantile(.75))& (customers_df['amount_spent']<=customers.quantile(.95)),'Preferreted',(np.where((customers_df['amount_spent']<customers.quantile(.95)),'VIP','Normal')))
```

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
df1 = orders['CustomerID']
df2= orders['amount_spent']
df3=orders['Country']
```

```python
d = pd.concat([df1,df2,df3], axis=1)
countries=d.groupby(['CustomerID','Country']).sum()
```

```python
dd =pd.DataFrame(countries)
dd
```

```python

dd['Type_customer']=np.where((dd>=countries.quantile(.75))& (dd<=countries.quantile(.95)),'Preferreted',(np.where((dd<countries.quantile(.95)),'VIP','Normal')))
```

```python
dd
```

```python
clients = dd[dd['Type_customer']== 'VIP']
```

```python
vips1 = pd.DataFrame(clients.groupby(['Country'])['Type_customer'].value_counts().sort_values(ascending = False))
```

```python
vips1
#El país con más clientes VIP es United Kingdom 
```

```python
#Q3

clients2 = dd[(dd['Type_customer']== 'VIP') | (dd['Type_customer']== 'Preferreted')]
clients2
```

```python
vips2 = pd.DataFrame(clients2.groupby(['Country'])['Type_customer'].count().sort_values(ascending = False))
```

```python
vips2
#El país con más clientes es Reino Unido 
```
