#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
is_hage = titanic.Age[titanic.Gender == 'male']
is_mage = titanic.Age[titanic.Gender == 'female']
is_hage.hist(bins = np.arange(0,80,5),label = 'Hombres',color='y')
is_mage.hist(bins = np.arange(0,80,5),label = 'Mujeres',color='m')
plt.title('Distribución de edades por género')
plt.xlabel('Edad')
plt.legend()
plt.show()
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here


plt.figure(figsize=(12,9))

plt.subplot(3,1,1)
is_hage = titanic.Age[titanic.Gender == 'male']
is_mage = titanic.Age[titanic.Gender == 'female']
is_hage.hist(bins =10,label = 'Hombres',color='y')
is_mage.hist(bins = 10,label = 'Mujeres',color='m')
plt.title('Edades por género (bin=10)')
#plt.xlabel('Edad')
plt.legend()


#Segundo subplot 
plt.subplot(3,1,2)
is_hage = titanic.Age[titanic.Gender == 'male']
is_mage = titanic.Age[titanic.Gender == 'female']
is_hage.hist(bins =20,label = 'Hombres',color='y')
is_mage.hist(bins = 20,label = 'Mujeres',color='m')
plt.title('Edades por género (bin=20)')
#plt.xlabel('Edad')
plt.legend()


#Tercer subplot 
plt.subplot(3,1,3)
is_hage = titanic.Age[titanic.Gender == 'male']
is_mage = titanic.Age[titanic.Gender == 'female']
is_hage.hist(bins =50,label = 'Hombres',color='y')
is_mage.hist(bins = 50,label = 'Mujeres',color='m')
plt.title('Distribución de edades por género (bin=50)')
plt.xlabel('Edad')
plt.legend()


#plt.show()


```

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here

plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(titanic.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', pad=12);
```
