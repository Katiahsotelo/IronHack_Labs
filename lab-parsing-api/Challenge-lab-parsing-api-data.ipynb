```python
datos=pd.DataFrame()
for i in range(0,42):
    url=f'https://rickandmortyapi.com/api/character/?page={i}'
    page=requests.get(url)
    page1 = page.json()
    page2 = page1['results']
    new = pd.json_normalize(page2)
    datos=datos.append(new,ignore_index=True)
             
                                          
```
